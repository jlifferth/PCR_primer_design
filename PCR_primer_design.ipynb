## Principles for primer design

<ul>
    <li>18-24 bp length</li>
    <li>40-60% GC content</li>
    <li>melting temperature 50°-60°C</li>
    <li>self complimentary</li>
    <li>Cross dimer</li>
</ul>

## Steps:

<ol>
    <li>Retrieve Nucleotide FASTA sequence from NCBI</li>
    <li>Design Primer in NCBI-Primer BLAST</li>
    <li>Primer specific site identification</li>
    <li>Primer validation and PCR Primer properties using Primer STAT</li>
</ol>
